**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
data = {
    "Name": ["Ali", "Reza", "Mahdi", "Hasan", "Hamed"],
    "Age": [23, 18, 19, 20, 22],
    "Student_ID": [2544, 3655, 4156, 7536, 9801],
    "Score": [58, 69, 84, 98, 73]
}

students_df = pd.DataFrame(data)

def assign_grade(score):
  if score >= 90:
    return 'A'
  elif score >= 80:
      return 'B'
  elif score >= 70:
      return 'C'
  else:
      return 'D'

students_df['Grade'] = students_df['Score'].apply(assign_grade)
print(students_df)
````

**Problem 11:** You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

# 1
total_sales = sales_df.sum()
print("*******************Q1*******************")
print("Total sales of each product:\n" ,total_sales)

# 2
avg_sales = sales_df.mean()
print("\n*******************Q2*******************")
print("Average monthly sales for each product:\n" ,avg_sales)

# 3
highest_sales_A = sales_df.loc[sales_df['Product A Sales'].idxmax()]
highest_sales_B = sales_df.loc[sales_df['Product B Sales'].idxmax()]
highest_sales_C = sales_df.loc[sales_df['Product C Sales'].idxmax()]
print("\n*******************Q3*******************")
print("highest sales for each product:")
print("Product A: ", highest_sales_A['Month'], "Sales:", highest_sales_A['Product A Sales'])
print("Product B: ", highest_sales_B['Month'], "Sales:", highest_sales_B['Product B Sales'])
print("Product C: ", highest_sales_C['Month'], "Sales:", highest_sales_C['Product C Sales'])

# 4
sales_df['Year'] = sales_df['Month'].dt.year
sales_in_2022 = sales_df[sales_df['Year'] == 2022]
percent_change_A = ((sales_in_2022['Product A Sales'].iloc[0] - sales_in_2022['Product A Sales'].iloc[-1]) / sales_in_2022['Product A Sales'].iloc[0]) * 100
percent_change_B = ((sales_in_2022['Product B Sales'].iloc[0] - sales_in_2022['Product B Sales'].iloc[-1]) / sales_in_2022['Product B Sales'].iloc[0]) * 100
percent_change_C = ((sales_in_2022['Product C Sales'].iloc[0] - sales_in_2022['Product C Sales'].iloc[-1]) / sales_in_2022['Product C Sales'].iloc[0]) * 100
print("\n*******************Q4*******************")
print("Product A:", percent_change_A, "%")
print("Product B:", percent_change_B, "%")
print("Product C:", percent_change_C, "%")

# 5
plt.figure(figsize=(12, 6))
plt.plot(sales_df['Month'], sales_df['Product A Sales'], label='Product A', marker='*', color = 'red')
plt.plot(sales_df['Month'], sales_df['Product B Sales'], label='Product B', marker='*', color = 'green')
plt.plot(sales_df['Month'], sales_df['Product C Sales'], label='Product C', marker='*', color = 'blue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Chart')
plt.legend()


# 6
correlation_AB = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print("\n*******************Q6*******************")
print("Correlation Coefficient:", correlation_AB)
plt.show()
````
